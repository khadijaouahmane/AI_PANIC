```python
import os, time, signal, subprocess, requests
from pyngrok import ngrok

PROJECT_DIR = "/content/drive/MyDrive/scream_dataset"
PORT = 5000

# 1) aller dans le dossier
os.chdir(PROJECT_DIR)

# 2) tuer tout ce qui utilise le port (proprement)
subprocess.run("fuser -k 5000/tcp >/dev/null 2>&1 || true", shell=True)
subprocess.run("pkill -f app.py >/dev/null 2>&1 || true", shell=True)

# 3) lancer Flask en arri√®re-plan
flask_proc = subprocess.Popen(
    ["python3", "-u", "app.py"],
    stdout=open("flask_log.txt", "w"),
    stderr=subprocess.STDOUT
)

# 4) attendre que le serveur r√©ponde
print("‚è≥ Attente d√©marrage Flask...")
ok = False
for _ in range(40):  # ~40s max
    try:
        r = requests.get(f"http://127.0.0.1:{PORT}/", timeout=1)
        if r.status_code == 200:
            ok = True
            break
    except:
        time.sleep(1)

if not ok:
    print("‚ùå Flask n‚Äôa pas d√©marr√©. Voici les logs :")
    print(open("flask_log.txt","r").read()[-2000:])
    raise SystemExit

print("‚úÖ Flask OK")

# 5) ouvrir tunnel ngrok
public_url = ngrok.connect(PORT, "http").public_url
print("\nüåç LIEN PUBLIC (ouvre-le dans ton navigateur) :")
print(public_url)

print("\n‚úÖ Tu peux maintenant :")
print(f"- Ouvrir le site : {public_url}/")
print(f"- API JSON : {public_url}/predict")
```
